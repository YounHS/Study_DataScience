```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
train = pd.read_csv('../Data/train.csv')
test = pd.read_csv('../Data/test.csv')
```

```python
train_test_data = [train, test]
print(train_test_data)
```

```python
def bar_chart(feature):
    surv = train[train['Survived'] == 1][feature].value_counts()
    dead = train[train['Survived'] == 0][feature].value_counts()
    df = pd.DataFrame([surv, dead])
    df.index = ['Survived', 'Dead']
    df.plot(kind='bar', stacked=True, figsize=(10, 5))
    plt.show()
```

```python
bar_chart('Sex')
```

```python
# Name은 데이터 분석에 있어 중요한 컬럼이 아님
train.drop('Name', axis=1, inplace=True)
test.drop('Name', axis=1, inplace=True)
```

```python
train.head()  # Name이 삭제됬음을 확인 가능
```

```python
# 성별 매핑 및 성별과 생존여부 관계
gender_mapping = {'male': 0, 'female': 1}

for dataSet in train_test_data:
    dataSet['Sex'] = dataSet['Sex'].map(gender_mapping)

bar_chart('Sex')
```

```python
# 나이와 생존여부 관계
# Age는 결측치가 존재하므로 결측치 대체 필요
train['Age'].fillna(train.groupby('Sex')['Age'].transform('median'), inplace=True)
test['Age'].fillna(train.groupby('Sex')['Age'].transform('median'), inplace=True)
```

```python
facet = sb.FacetGrid(train, hue='Survived', aspect=4)
facet.map(sb.kdeplot, 'Age', shade=True)
facet.set(xlim=(0, train['Age'].max()))
facet.add_legend()
sb.axes_style('darkgrid')
```

```python
# Age의 범위는 꽤 넓으므로 Feature Engineering의 Binning 기술을 사용
for dataSet in train_test_data:
    dataSet.loc[dataSet['Age'] <= 16, 'Age'] = 0,
    dataSet.loc[(dataSet['Age'] > 16) & (dataSet['Age'] <= 26), 'Age'] = 1,
    dataSet.loc[(dataSet['Age'] > 26) & (dataSet['Age'] <= 36), 'Age'] = 2,
    dataSet.loc[(dataSet['Age'] > 36) & (dataSet['Age'] <= 62), 'Age'] = 3,
    dataSet.loc[dataSet['Age'] > 62, 'Age'] = 4
    
train.head()
```

```python
bar_chart('Age')
```

```python
Pclass1 = train[train['Pclass'] == 1]['Embarked'].value_counts()
Pclass2 = train[train['Pclass'] == 2]['Embarked'].value_counts()
Pclass3 = train[train['Pclass'] == 3]['Embarked'].value_counts()
df = pd.DataFrame([Pclass1, Pclass2, Pclass3])
df.index = ['1st class', '2nd class', '3rd class']
df.plot(kind='bar', stacked=True, figsize=(10, 5))

plt.show()
```

```python
# ML classfier 작업을 위해 텍스트 데이터 매핑
embarked_mapping = {'S': 0, 'C': 1, 'Q': 2}

for dataSet in train_test_data:
    dataSet['Embarked'] = dataSet['Embarked'].map(embarked_mapping)
    
# S 값이 1,2,3 등급의 다수를 차지하므로 결측치는 S로 대체하기로 함
train['Embarked'].fillna('S')

# 티켓 가격은 좌석 등급(클래스)과 관련성이 높으며 클래스는 결측치 값이 없으므로 클래스의 가격 평균을 결측치 값으로 대체
train['Fare'].fillna(train.groupby('Pclass')['Fare'].transform('median'), inplace=True)
test['Fare'].fillna(test.groupby('Pclass')['Fare'].transform('median'), inplace=True)
```

```python
facet = sb.FacetGrid(train, hue='Survived', aspect=4)
facet.map(sb.kdeplot, 'Fare', shade=True)
facet.set(xlim=(0, train['Fare'].max()))
facet.add_legend()
plt.show()
```

```python
for dataSet in train_test_data:
    dataSet.loc[dataSet['Fare'] <= 17, 'Fare'] = 0,
    dataSet.loc[(dataSet['Fare'] > 17) & (dataSet['Fare'] <= 30), 'Fare'] = 1,
    dataSet.loc[(dataSet['Fare'] > 30) & (dataSet['Fare'] <= 100), 'Fare'] == 2,
    dataSet.loc[dataSet['Fare'] > 100, 'Fare'] = 3
    
train.head()
```

```python
for dataSet in train_test_data:
    dataSet['Cabin'] = dataSet['Cabin'].str[:1]

Pclass_1 = train[train['Pclass'] == 1]['Cabin'].value_counts()
Pclass_2 = train[train['Pclass'] == 2]['Cabin'].value_counts()
Pclass_3 = train[train['Pclass'] == 3]['Cabin'].value_counts()
df = pd.DataFrame([Pclass_1, Pclass_2, Pclass_3])
df.index = ['1st class', '2nd class', '3rd class']
df.plot(kind='bar', stacked=True, figsize=(10, 5))

plt.show()
```

```python
# classifier를 위해 Cabin 데이터 binning
cabin_mapping = {'A': 0, 'B': 0.4, 'C': 0.8, 'D': 1.2, 'E': 1.6, 'F': 2, 'G': 2.4, 'T': 2.8}

for dataSet in train_test_data:
    dataSet['Cabin'] = dataSet['Cabin'].map(cabin_mapping)
```

```python
# Cabin 결측치는 클래스와 관련있기 때문에 각 클래스별 중간값을 결측치 값으로 대체
train['Cabin'].fillna(train.groupby('Pclass')['Cabin'].transform('median'), inplace=True)
test['Cabin'].fillna(test.groupby('Pclass')['Cabin'].transform('median'), inplace=True)
```

```python
# 가족 규모에 따른 생존/사망률
train['FamilySize'] = train['SibSp'] + train['Parch'] + 1
test['FamilySize'] = test['SibSp'] + test['Parch'] + 1
```

```python
# 탑승 그룹 규모에 따른 생존여부 관계
facet = sb.FacetGrid(train, hue='Survived', aspect=4)
facet.map(sb.kdeplot, 'FamilySize', shade=True)
facet.set(xlim=(0, train['FamilySize'].max()))
facet.add_legend()

plt.show()  # 혼자 탑승 시, 사망률 높음
```

```python
# 가족 규모를 binning
family_mapping = {1: 0, 2: 0.4, 3: 0.8, 4: 1.2, 5: 1.6, 6: 2, 7: 2.4, 8: 2.8, 9: 3.2, 10: 3.6, 11: 4}

for dataSet in train_test_data:
    dataSet['FamilySize'] = dataSet['FamilySize'].map(family_mapping)
    
train.head()
```

```python
# 필요없는 데이터 한꺼번에 삭제
drop_datas = ['Ticket', 'SibSp', 'Parch']
train = train.drop(drop_datas, axis=1)
test = test.drop(drop_datas, axis=1)

train = train.drop('PassengerId', axis=1)
```

```python
print(train.shape, test.shape)
```

```python
train = pd.get_dummies(train)
train.shape
```

```python
test = pd.get_dummies(test)
test.shape
```

```python
train_label = train['Survived']
train_data = train.drop('Survived', axis=1)
test_data = test.drop('PassengerId', axis=1).copy()
```

```python
print(train.shape, test.shape)
```

```python
train_data.info()
```

```python
test_data.info()
```

```python
train_data['Embarked'] = train_data['Embarked'].fillna(0.0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
import pandas as pd

from sklearn.utils import shuffle
```

```python
# 학습시키기 전, 주어진 데이터가 정렬되어 있어 학습에 방해가 될 수 있으므로 셔플링
train_data, train_label = shuffle(train_data, train_label, random_state=5)
```

```python
# 모델 학습과 평가에 대한 파이프라인 생성
# scikit-learn에서 제공하는 fit(), predict()를 사용하면 매우 간단하게 학습과 예측 가능
def predictionModel(model):
    model.fit(train_data, train_label)
    prediction = model.predict(test_data)
    accuracy = round(model.score(train_data, train_label) * 100, 2)
    print('Accuracy: ', accuracy, '%')
    return prediction
```

```python
# Logistic Regression
log_pred = predictionModel(LogisticRegression())
# SVM
svm_pred = predictionModel(SVC())
# kNN
knn_pred_4 = predictionModel(KNeighborsClassifier(n_neighbors=4))
# Random Forest
rf_pred = predictionModel(RandomForestClassifier(n_estimators=100))
# Navie Bayes
nb_pred = predictionModel(GaussianNB())
```

```python
submission = pd.DataFrame({
    "PassengerId": test['PassengerId'],
    "Survived": rf_pred
})
```
