```python
# datetime: 시간. 연-월-일 시:분:초 로 표현
# season: 계절. 봄(1), 여름(2), 가을(3), 겨울(4)
# holiday: 공휴일. 공휴일(1), 공휴일X(0)
# workingday: 근무일. 근무일(1), 근무일X(0)
# weather: 날씨
#          1: 아주 깨끗한 날씨, 또는 아주 약간의 구름
#          2: 약간의 안개와 구름이 끼어있는 날씨
#          3: 약간의 눈, 비가 오거나 천둥
#          4: 아주 많은 비가 오거나 우박
# temp: 온도. 섭씨
# atemp: 체감온도
# humidity: 습도
# windspeed: 풍속
# casual: 비회원의 자전거 대여량
# registered: 회원의 자전거 대여량
# count: 총 자전거 대여량, 비회원+회원과 동일
```
