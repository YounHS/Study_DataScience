```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np

from scipy import stats
from scipy.stats import skew, norm
```

```python
train = pd.read_csv('../Data/train.csv')
test = pd.read_csv('../Data/test.csv')

print(train.shape, test.shape)
```

```python
# Id 중복확인
idsUnique = len(set(train.Id))
idsUnique
```

```python
# Id 제거
Id_test = test.Id  # test의 id는 캐글 제출 시 필요하므로 keep
train.drop('Id', axis = 1, inplace = True)
test.drop('Id', axis=1, inplace = True)
```

```python
# Preprocessing
```

```python
# GrLivArea와 SalePrice 간 산점도 확인
GLA = 'GrLivArea'
data = pd.concat([train['SalePrice'], train[GLA]], axis=1)
data.plot.scatter(x=GLA, y='SalePrice', ylim=(0,800000))

# 상기 결과 오른쪽 하단의 두 개의 큰 집이 싼 값에 팔렸고, 이는 이상치라고 생각되니 제거
train = train[(train[GLA] < 4000) | (train['SalePrice'] > 700000)]

data = pd.concat([train['SalePrice'], train[GLA]], axis=1)
data.plot.scatter(x=GLA, y='SalePrice', ylim=(0,800000))
```

```python
sb.distplot(train['SalePrice'], fit=norm)
fig = plt.figure()

# 왼쪽으로 치우친 왜도가 보이고, 로그 표준화하여 해결한다.
train['SalePrice'] = np.log1p(train['SalePrice'])
y = train['SalePrice']

sb.distplot(train['SalePrice'], fit=norm)
fig = plt.figure()
```
